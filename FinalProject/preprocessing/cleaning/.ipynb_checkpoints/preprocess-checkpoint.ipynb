## One-hut encode

```python
#Using a hashmap, we will encode

def one_hut_encode(hashmap, origFeature):
    #Create New Features
    for key in hashmap.keys():
        newDF[key] = np.zeros(len(newDF))
        
    #Categorize Rows into these features
    for idx, row in newDF.iterrows():
        currentFeature = row[origFeature] #original input for the old feature
        for key in hashmap.keys():
            if currentFeature in hashmap.get(key):
                newDF.loc[newDF.index == idx, key] = 1
    
    #Delete the old column
    del newDF[origFeature]
```

```python
''' 
 Rows with the same “id” column value will be identified. 
 Rows with identical “id” column values will have their ages to be nan or Null. 
 These rows will then be merged into 1 row by using the average 
 of their other features as the new values for the merged row.
'''

dfColumns = newDF.columns
newDF2 = pd.DataFrame(index=[], columns=dfColumns) #Cleaned dataframe from sample

#Setting new index
newDF.set_index('id', inplace=True)

array = []
dfArray = []
restrictedVariables = ['Gender', 'Customer Type', 'Type of Travel', 'Class']

#Main operation
#Selecting row
for index, row in newDF.iterrows():
    if index in array:
        pass
    else:
        array.append(index)
        dup = newDF[newDF.index == index]
        dfArray.append(dup)
        

for df in dfArray:
    if len(df) > 1:    
        rowHashmap = {}
        index = df.index[0]
        for variable in df:
            if variable not in restrictedVariables:
                #Get Mean of the Variable and append to hashmap
                value = ceil(df[variable].mean())
                rowHashmap.update({'id' : index, variable : [value]})

        #Make new Dataframe row
        print(f'DF of {index}: {rowHashmap}')
        converted = pd.DataFrame.from_dict(rowHashmap)
        conv = pd.DataFrame(rowHashmap, index=[index])
        newDF2 = pd.concat([newDF2, conv])
    else:
        pass
    
if len(newDF2) == 0:
    newDF2 = newDF
else:
    #Remove all rows with duplicate "id" value
    newDF = newDF.drop_duplicates(subset='id', keep = False)
    
    #concatenate the two Dataframes
    newDF2 = pd.concat([newDF2, newDF])
    

print(f'Length of new Dataframe: {len(newDF2)}')
newDF2.describe()
```

```python
#Setting "id" index back as a column
newDF2['id'] = newDF2.index
del newDF2['index']
```

```python
print(newDF2['satisfaction'].unique())
newDF2['id'].count()
```

```python
# Data Normalization for all continuous data

contin = []
print('Continuous Data')
for c in newDF2.columns:
    restricted = ['id']
    condition = len(newDF2[newDF2[c] > 5]) > 1
    if condition == True and c not in restricted:
        contin.append(c)
        print(f'Column: {c}')

for column in contin:
    newDF2[column] = (newDF2[column] - newDF2[column].min()) / (newDF2[column].max() - newDF2[column].min())
```
