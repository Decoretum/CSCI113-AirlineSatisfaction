```python
#Dataset
# Since K Cross Fold Validation hasn’t been executed yet in this stage, 
# then the two datasets (training, testing) will be merged into 1 dataset.

currentPath = os.getcwd() #Current working directory
testPath = currentPath.replace('FinalProject/preprocessing/cleaning','FinalProject/dataset/test.csv')
trainPath = currentPath.replace('FinalProject/preprocessing/cleaning','FinalProject/dataset/train.csv')

df = pd.read_csv(testPath)
df2 = pd.read_csv(trainPath)
finalDF = pd.concat([df, df2])
finalDF.describe()
```

```python
# Remove the “Unnamed, 0” column from the dataset.
finalDF = finalDF.drop(columns = ['Unnamed: 0'])
finalDF.describe()
```

```python
# Convert rows’ “Age” column’s data type into an integer.
newDF = finalDF.astype({"Age" : int})
newDF['Age'].dtypes
```

```python
# Rows with an Age value of 0 or with a float type will be removed from the dataset.

newDF = finalDF[(finalDF['Age'] > 0) & (type(finalDF['Age']) != 'float')]
newDF = newDF.astype({'Age' : int})
newDF['Age']
```

```python
# Rows with nan or Null values in any of their column will be removed from the dataset.



print(f'Rows Before dropping NaN or Null values: {len(newDF)}')

newDF.dropna(axis = 0, how='any', inplace = True)

print(f'Rows after dropping NaN or Null values: {len(newDF)}')
```

```python
''' 
 Rows with the same “id” column value will be identified. 
 Rows with identical “id” column values will have their ages to be nan or Null. 
 These rows will then be merged into 1 row by using the average 
 of their other features as the new values for the merged row.
'''

dfColumns = newDF.columns
newDF2 = pd.DataFrame(index=[], columns=dfColumns) #Cleaned dataframe from sample

#Setting new index
newDF.set_index('id', inplace=True)

array = []
dfArray = []
restrictedVariables = ['Gender', 'Customer Type', 'Type of Travel', 'Class']

#Main operation
#Selecting row
for index, row in newDF.iterrows():
    if index in array:
        pass
    else:
        array.append(index)
        dup = newDF[newDF.index == index]
        dfArray.append(dup)
        


for df in dfArray:
    if len(df) > 1:    
        rowHashmap = {}
        index = df.index[0]
        for variable in df:
            #Get Mean of the Variable and append to hashmap
            value = ceil(df[variable].mean())
            rowHashmap.update({'id' : index, variable : [value]})

        #Make new Dataframe row
        print(f'DF of {index}: {rowHashmap}')
        converted = pd.DataFrame.from_dict(rowHashmap)
        conv = pd.DataFrame(rowHashmap, index=[index])
        newDF2 = pd.concat([newDF2, conv])
    else:
        pass
    
if len(newDF2) == 0:
    newDF2 = newDF
else:
    #Remove all rows with duplicate "id" value
    newDF = newDF.drop_duplicates(subset='id', keep = False)
    
    #concatenate the two Dataframes
    newDF2 = pd.concat([newDF2, newDF])
    

print(f'Length of new Dataframe: {len(newDF2)}')
newDF2.describe()
    
```
