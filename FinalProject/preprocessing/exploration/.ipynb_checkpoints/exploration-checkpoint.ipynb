```python
contin = []
print('Continuous Data')

def conditioning_cont(df, column):
    c = len(df[(df[column] <= 1) & (df[column] > -1) & (df[column] != 0)])
    if c == 0:
        return False
    else:
        return True
    
for c in df.columns:
    restricted = ['id']
    if conditioning_cont(df, c) == True and c not in restricted:
        contin.append(c)
        print(f'Column: {c}')
```

```python
df['Flight Distance'].max() == 5
```

```python
categ = []
print('Categorical Data')

def conditioning_categ(df, column):
    binary = [0, 1]
    c = df[column].max() == 5
    if c == True:
        return True
    else:
        return False
    
for c in df.columns:
    restricted = ['id']
    if conditioning_categ(df, c) == True and c not in restricted:
        categ.append(c)
        print(f'Column: {c}')
```

## Bivariate Analysis
For continuous variables, develop scatterplots and lineplots between the continuous variable in relation to the â€œsatisfactionâ€ variable.

Develop grouped column charts between categorical variables and the â€œsatisfactionâ€ variable

Develop grouped column charts between nominal variables and the â€œsatisfactionâ€ variable

```python
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['satisfaction'] == 0]['Flight Distance'], df[df['satisfaction'] == 1]['Flight Distance']],
            labels=['Unsatisfied', 'Satisfied'])
plt.title('Box Plot of Satisfaction and Flight Distance')
plt.xlabel('Satisfaction')
plt.ylabel('Flight Distance')
plt.grid(True)
plt.show()
```

```python
# Change satisfaction values from 0 and 1 to "Unsatisfied" and "Satisfied" respectively
df['satisfaction'] = df['satisfaction'].replace({0: 'Unsatisfied', 1: 'Satisfied'})

# Specify the categorical variable
categ = ['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'Leg room service', 'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness']

for col in categ:
    filtered_df = df[df[col] != 0]

    # Group by the categorical variable and satisfaction, then unstack to pivot the table
    grouped = filtered_df.groupby([col, 'satisfaction']).size().unstack()

    # Plot the grouped column chart
    grouped.plot(kind='bar', stacked=False)

    plt.title(f'Column Chart for Satisfaction and {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
